 ## 6. 2015년 이전의 데이터를 합하여 '2015년 이전' 컬럼을 만들고, 최근 3개년간의 데이터를 합하여 '최근 3개년' 컬럼을 만든 후, '최근 증가율' 컬럼을 만든다.

```python
cctv.fillna(0)
```

```python
cctv = pd.to_numeric(cctv,downcast='integer')
cctv
```

```python
cctv [['2011년 이전', '2012년','2013년']]= \
    cctv[['2011년 이전', '2012년','2013년']].fillna(0).astype(int)
cctv
```

```python
cctv['2015년 이전'] = cctv['2011년 이전'] + cctv['2012년'] + cctv['2013년'] + cctv['2014년'] + cctv['2015년']
cctv
```

```python
cctv['최근 3개년'] = cctv[['2016년','2017년','2018년']].sum(axis=1)
cctv
```

```python
#cctv['최근 3개년'] = cctv.lioc[:,10:11].sum(axis=1)
cctv['최근 증가율'] = cctv['최근 3개년'] / cctv['2015년 이전'] *100
cctv.head()
```

 ## 8. 서울시 인구 데이터와 cctv 데이터를 합치고 인데스를 '자치구'로 만든다.

```python
pop.head()
```

```python
df = pd.merge(cctv, pop, on='자치구')
df.head()
```

```python
df.set_index('자치구',inplace=True)
df.head()
```

 ## 9. CCTV 현황을 다양한 형태로 시각화하여 보여준다.

```python
#basic
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (14,10)
plt.rcParams["font.size"] = 14
```

 ##### 자치구별 CCTV 댓수 (내림차순으로)

```python
df['소계'].sort_values().plot(kind='barh',grid=True)
plt.title('CCTV 증가 대수')
plt.show()
```

```python
df['인당 CCTV'] = df['소계'] / df['인구수']
df['인당 CCTV'].sort_values().plot(kind='barh', grid=True)
plt.title('인당 CCTV 대수')
plt.show()
```

 ### - 자치구별 최근 증가율 (내림차순으로)

```python
df['최근 증가율'].sort_values().plot(kind='barh',grid=True)
plt.title('CCTV 최근 증가율')
plt.show()
```

 ### -인구수와 CCTV 댓수의 선점도

```python
plt.scatter(df['인구수'],df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.grid()
plt.show()
```

 ##### 인구수와 CCTV 댓수의 산점도 및 회귀선

```python
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000,700000,101)
print(f1)
print()
print(fx)
```

```python
plt.scatter(df['인구수'],df['소계'], s=50)
plt.plot(fx,f1(fx),ls = 'dashed', lw = 3, color= 'r')
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.grid()
plt.show()
```

 ### -인구수와 CCT 댓수의 선점도, 회귀선 및 자치구명

```python
df.head()
```

```python
df['오차'] = np.abs(df['소계'] - f1(df['인구수']))
df.head()
```

```python
df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head(10)
```

```python
plt.scatter(df['인구수'],df['소계'],c=df['오차'], s=50)
plt.plot(fx,f1(fx),ls = 'dashed', lw = 3, color= 'm')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당 비율')
plt.colorbar()
plt.grid()
plt.show()
```
