```python
# 12- 4
xdata=['안지영', '홍지수', '황예린']
ydata1= [90, 85, 88]
ydata2= [83, 88, 91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
plt.legend(loc='upper center')
plt.title('세 명 학생의 국어, 영어 성적')

plt.show()
```

```python
xdata=['안지영', '홍지수', '황예린']
ydata1= [90, 85, 88]
ydata2= [83, 88, 91]
plt.plot(xdata, ydata1, 'ro-', label='국어')
plt.plot(xdata, ydata2, 'cx-', label='영어')
plt.legend(loc='upper center')
plt.title('세 명 학생의 국어, 영어 성적')

plt.show()
```

```python
# 12-6
xdata = list(range(30))
ydata = []

for x in xdata:
    y =2 *x
    ydata.append(y)
    
plt.plot(xdata, ydata, label='y=2x', color = 'magenta')
plt.xlim(-2,35)
plt.ylim(-2,70)
plt.xticks(list(range(0,35,10)))
plt.yticks(list(range(0,70,20)))

plt.grid(True)
plt.show()
```

```python
# sigmoid 함수

%matplotlib inline
xs = np.linspace(-3,3, 601)
ys = 1 / (1 + np.exp(-xs))
plt.plot(xs,ys)
plt.title('시그모이드 그래프')
plt.grid()
plt.show()
```
